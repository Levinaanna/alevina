```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
X = [1,   1.5, 2,   3, 5, 6]
Y = [1.8, 1,   3.6, 2, 1, 3]

plt.plot(X, Y) 
plt.show()
```

```python
plt.scatter(X,Y)  
plt.title("точки")
plt.xlabel("Ось A")
plt.ylabel("Ось B")
plt.show()
```

```python
plt.scatter(X, Y)
plt.xlim(0,5)
plt.ylim(0,2)
plt.show()
```

```python
plt.scatter(X, Y, s=[20, 60, 30, 100, 60, 10], c='red', marker='^')
plt.show()
```

```python

size = [60, 70, 80, 90, 100, 130]
colors = ['yellow']
plt.scatter(X, Y, s=size, c=colors, marker='^')
plt.show()
```

```python
markers = ['o', '*', '8', 'D', '^', 'P']

for x, y, m in zip(X, Y, markers):
    plt.scatter([x],[y], marker=m, s=100)
plt.show()
```

```python
markers = ['*', '*', '*', '*',  '*']
for x, y, m in zip(X, Y, markers):
    plt.scatter([x],[y], marker=m, s=100)
plt.show()
```

```python
dots = ['Первая', "Вторая", "Третья", "Четвертая", "Пятая", "Шестая"]

for x, y, m, d in zip(X, Y, markers, dots):
    plt.scatter(x, y, marker=m, s=100)
    plt.text(x+0.2, y+0.4, d) # +0.1 - это чтобы текст не наползал на маркер, а отрисовывался чуть выше и правее 
plt.show()
```

```python
text= open('alice.txt', 'r', encoding='utf-8')
text1 = text.read()
text2=[w for w in text1.split() if w.isalnum()]
print(text2)
print(len(text2))
```

```python
text3 = ' '.join(text2)
print(text3)
print(len(text3))
```

```python
from nltk.tokenize import word_tokenize
print(word_tokenize(text3))
text4=word_tokenize(text3)
```

```python
from nltk.corpus import stopwords
sw = stopwords.words('english')
filtered = [w.lower() for w in text4 if w.lower() not in sw]
print(filtered)
print(len(filtered))
```

```python
from collections import Counter 
final = Counter(filtered)
print(final.most_common(30))
```

```python
import matplotlib
X = range(30)
Y = [word[1] for word in final.most_common (30)]

plt.bar(X, Y)
plt.show()
```
