```python
# Задание 1
#Записываю текст в переменную и разбиваю на слова с помощью .split()
text="""варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг"""
monograms=text.split()
print(monograms)
```

```python
#Создаю список биграмм при помощи цикла for. i-переменная, одна из чатей биграммы. 
bigrams=[]
for i in range(len(monograms)-1):
    bigrams.append([monograms[i],monograms[i+1]])#При помощи .append добавляю вторую чать биграммы и записываю их.
print (bigrams)    
```
